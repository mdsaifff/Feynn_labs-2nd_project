```python
import pandas as pd
import numpy as np
import seaborn as sns
import datetime as dt
import matplotlib.pyplot as plt
from sklearn.preprocessing import OrdinalEncoder
from mlxtend.preprocessing import minmax_scaling
from sklearn.preprocessing import minmax_scale
from sklearn.decomposition import PCA
```

```python
orig_data= pd.read_csv("/content/new_magicbricks 5.csv")
orig_data
```

```python
orig_data.isnull().sum()
```

```python
data=orig_data.copy()
```

```python
#ordinal encoding
o_encode1= OrdinalEncoder(categories=[['UnderConstruction','NewPropertyGetDocumentsVerified','ResaleGetDocumentsVerified','ReadytoMove']])
data['status']= o_encode1.fit_transform(orig_data[['status']])
data
```

```python
for id in data['rera-id']:
  if id is 'NaN':
    data['verified_id']=0
  else:
    data['verified_id']=1  
```

```python
data.drop(columns=['name','rera-id'])
```

```python
for category in data['recommended-for']:
  if category is 'Retirees,,Family,,Couple,,':
    data['max_capacity']=4
  elif category is 'Family,,Couple,,SingleProfessionals,,':
    data['max_capacity']=3
  elif category is 'Family,,Couple,,Students,,':
    data['max_capacity']=6
  elif category is 'Retirees,,Family,,Students,,':
    data['max_capacity']=5
  elif category is 'Family,,':
    data['max_capacity']=4
  else:
    data['max_capacity']=2
```

```python
data.drop(['name','rera-id'], axis = 1)
```

```python
data['water-availability'].value_counts()
```

```python
data['water-availability']= orig_data['water-availability'].map({'44HoursAvailable': 1, 'No/RarePowercut': 2})
```

```python
data['water-availability']= data['water-availability'].fillna(0)
data['water-availability'].unique()
```

```python
data['status-of-electricity'].value_counts()
```

```python
data['status-of-electricity']= orig_data['status-of-electricity'].map({'44HoursAvailable': 1, 'No/RarePowercut': 2})
```

```python
data['status-of-electricity']= data['status-of-electricity'].fillna(0)
data['status-of-electricity'].unique()
```

```python
data=data.drop(columns=['developer',])
data
```

```python
data['furnishing']= orig_data['furnishing'].map({'Semi-Furnished': 2, 'Unfurnished': 1})
data['furnishing']= data['furnishing'].fillna(0)
data['furnishing'].unique()
```
